```python
car_sales.info()
```

```python
car_sales['Total Sales'] = car_sales['Price'].cumsum()
car_sales
```

# Hard Disease Visualization

<img src="matplotlib-anatomy.png" />`

```python
heart_disease = pd.read_csv('heart-disease.csv')
heart_disease
```

```python
heart_disease['age'].plot.hist(bins = 20);
```

```python
heart_disease.plot.hist(figsize=(10, 30), subplots=True);
```

```python
heart_disease_over_50 = heart_disease[heart_disease['age'] > 50]
heart_disease_over_50
```

```python
heart_disease_over_50.plot(
    kind='scatter',
    x='age',
    y='chol',
    c='target'
);
```

```python
fig, ax = plt.subplots()
heart_disease_over_50.plot(
    kind='scatter',
    x='age',
    y='chol',
    c='target',
);
```

```python
fig, ax = plt.subplots()
heart_disease_over_50.plot(
    kind='scatter',
    x='age',
    y='chol',
    c='target',
    ax=ax,
    figsize=(10, 5)
);
```

```python
fig, ax = plt.subplots(figsize=(10, 5))

scatter = ax.scatter(
    x = heart_disease_over_50['age'],
    y = heart_disease_over_50['chol'],
    c = heart_disease_over_50['target'],
)


ax.set(
    title='Heart Disease & Cholesterol',
    xlabel='Age',
    ylabel='Cholesterol'
);

ax.legend(*scatter.legend_elements(), title='Target');

ax.axhline(
    heart_disease_over_50['chol'].mean(),
    linestyle=':',
    color='red'
);
```

### OO method (Above) is preferred above the regular stateless method (Ones more above)

```python
plt.style.use('seaborn')

fig, (ax1, ax2) = plt.subplots(
    figsize=(10, 10),
    nrows=2,
    ncols=1,
    sharex=True
)

scatter = ax1.scatter(
    x=heart_disease_over_50['age'],
    y=heart_disease_over_50['chol'],
    c=heart_disease_over_50['target'],
);

scatter2 = ax2.scatter(
    x=heart_disease_over_50['age'],
    y=heart_disease_over_50['thalach'],
    c=heart_disease_over_50['target'],
);

ax1.set(
    title='Heart Disease & Cholesterol',
#     xlabel='Age',
    ylabel='Cholesterol'
);

ax2.set(
    title='Heart Disease & Cholesterol',
    xlabel='Age',
    ylabel='Max Heartrate'
);

ax1.legend(*scatter.legend_elements(), title='Target');
ax2.legend(*scatter2.legend_elements(), title='Target');

ax1.axhline(
    y=heart_disease_over_50['chol'].mean(),
    color='red',
    linestyle='--'
);

ax2.axhline(
    y=heart_disease_over_50['thalach'].mean(),
    color='black',
    linestyle=':'
);

fig.suptitle('Heart Disease Analysis', size=20, weight='bold');
```

```python
plt.style.use('seaborn')

fig, (ax1, ax2) = plt.subplots(
    figsize=(10, 10),
    nrows=2,
    ncols=1,
    sharex=True
)

scatter = ax1.scatter(
    x=heart_disease_over_50['age'],
    y=heart_disease_over_50['chol'],
    c=heart_disease_over_50['target'],
    cmap='winter'
);

scatter2 = ax2.scatter(
    x=heart_disease_over_50['age'],
    y=heart_disease_over_50['thalach'],
    c=heart_disease_over_50['target'],
    cmap='summer'
);

ax1.set(
    title='Heart Disease & Cholesterol',
#     xlabel='Age',
    ylabel='Cholesterol'
);

ax1.set_xlim([50, 80])

ax2.set(
    title='Heart Disease & Cholesterol',
    xlabel='Age',
    ylabel='Max Heartrate'
);

ax1.legend(*scatter.legend_elements(), title='Target');
ax2.legend(*scatter2.legend_elements(), title='Target');

ax1.axhline(
    y=heart_disease_over_50['chol'].mean(),
    color='red',
    linestyle='--'
);

ax2.axhline(
    y=heart_disease_over_50['thalach'].mean(),
    color='black',
    linestyle=':'
);

fig.suptitle('Heart Disease Analysis', size=20, weight='bold');
```
