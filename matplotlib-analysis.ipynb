# Plot Pandas DataFrame

```python
car_sales = pd.read_csv('car-sales.csv')
```

```python
car_sales
```

```python
car_sales['Price'] = car_sales['Price'].str.replace('[\$\,\.]', '')
car_sales['Price'] = car_sales['Price'].str[:-2]
```

```python
car_sales
```

```python
car_sales['Sales Date'] = pd.date_range('8/8/2022', periods=len(car_sales))
```

```python
car_sales
```

```python
car_sales['Price'] = car_sales['Price'].astype(int)
```

```python
car_sales.info()
```

```python
car_sales['Total Sales'] = car_sales['Price'].cumsum()
car_sales
```

```python
car_sales.plot(x = 'Sales Date', y = 'Total Sales');
```

```python
car_sales.plot(x = 'Odometer (KM)', y = 'Price', kind = 'scatter');
```

```python
car_sales.plot(x = 'Odometer (KM)', y = 'Price', kind = 'bar');
```

```python
car_sales.plot.bar();
```

```python
car_sales.plot.hist();
```
