```python
import numpy as np
x = np.linspace(0, 10, 100)
x
```

```python
fig, ax = plt.subplots()
ax.plot(x, x**2);
```

```python
fig, ax = plt.subplots()
ax.scatter(x, np.exp(x));
```

```python
fig, ax = plt.subplots()
ax.scatter(x, np.sin(x));
```

```python
fig, ax = plt.subplots()
prices = { "Cookie": 10, "Butter": 8, "Chicken": 20 }
ax.set(title="Store Prices", ylabel="Products")
ax.bar(prices.keys(), prices.values());
```

```python
fig, ax = plt.subplots()
ax.barh(list(prices.keys()), prices.values());
```

```python
fig, ax = plt.subplots()
ax.hist(np.random.randn(1000));
```

```python
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(
    nrows=2,
    ncols=2,
    figsize=(10, 10)
)

ax1.plot(x, x/2);
ax2.scatter(np.random.random(10), np.random.random(10));
ax3.bar(prices.keys(), prices.values());
ax4.hist(np.random.randn(1000));
```

```python
fig, ((ax1, ax2, ax3), (ax4, ax5, ax6)) = plt.subplots(
    nrows=3,
    ncols=3,
    figsize=(10, 10)
)

ax1.plot(x, x/2);
ax2.scatter(np.random.random(10), np.random.random(10));
ax3.bar(prices.keys(), prices.values());
ax4.hist(np.random.randn(1000));
```
